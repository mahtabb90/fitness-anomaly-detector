# Dataset 2 UMAP Analysis – Mahtab

UMAP (Uniform Manifold Approximation and Projection) is a non-linear dimensionality reduction technique.

Unlike PCA, UMAP can capture complex patterns and structures in the data.

The goal is to visualize the dataset in 2D and explore hidden structures.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
import umap
```

```python
df = pd.read_csv("../data/gym_dataset.csv")
```

```python
umap_features = [
    "Session_Duration (hours)",
    "Avg_BPM",
    "Calories_Burned",
    "Weight (kg)",
    "BMI",
    "Fat_Percentage"
]

X = df[umap_features].copy().dropna().reset_index(drop=True)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
umap_model = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42)

X_umap = umap_model.fit_transform(X_scaled)

umap_df = pd.DataFrame(X_umap, columns=["UMAP1", "UMAP2"])

display(umap_df.head())
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=umap_df,
    x="UMAP1",
    y="UMAP2",
    hue=X["Calories_Burned"],
    palette="viridis"
)

plt.title("UMAP projection colored by Calories Burned")
plt.show()
```

```python
umap_model = umap.UMAP(n_neighbors=5, min_dist=0.0, random_state=42)

X_umap = umap_model.fit_transform(X_scaled)

umap_df = pd.DataFrame(X_umap, columns=["UMAP1", "UMAP2"])

plt.figure(figsize=(8,6))
sns.scatterplot(
    data=umap_df,
    x="UMAP1",
    y="UMAP2",
    hue=X["Calories_Burned"],
    palette="viridis"
)

plt.title("UMAP: n_neighbors=5, min_dist=0.0")
plt.show()
```

```python
umap_model = umap.UMAP(n_neighbors=50, min_dist=0.5, random_state=42)

X_umap = umap_model.fit_transform(X_scaled)

umap_df = pd.DataFrame(X_umap, columns=["UMAP1", "UMAP2"])

plt.figure(figsize=(8,6))
sns.scatterplot(
    data=umap_df,
    x="UMAP1",
    y="UMAP2",
    hue=X["Calories_Burned"],
    palette="viridis"
)

plt.title("UMAP: n_neighbors=50, min_dist=0.5")
plt.show()
```

```python
umap_model = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42)

X_umap = umap_model.fit_transform(X_scaled)

umap_df = pd.DataFrame(X_umap, columns=["UMAP1", "UMAP2"])

plt.figure(figsize=(8,6))
sns.scatterplot(
    data=umap_df,
    x="UMAP1",
    y="UMAP2",
    hue=X["Calories_Burned"],
    palette="viridis"
)

plt.title("UMAP: n_neighbors=15, min_dist=0.1")
plt.show()
```

## UMAP Parameter Tuning

Different parameter settings produce different structures:

- Low n_neighbors → more local clusters
- High n_neighbors → more global structure
- Low min_dist → tighter clusters
- High min_dist → more spread out data

This shows how UMAP can reveal different aspects of the data.

## Interpretation

UMAP reveals non-linear structures in the data.

Compared to PCA, UMAP shows more complex patterns and possible clusters.

Coloring by calories burned helps identify whether workout intensity is reflected in the structure.
